# Summary of Daily Minimum Wage Rate in the Philippines (2022)
*Source: Philippine Statistics Authority*


* <p align=justify> The following data provides information on the current daily minimum wage rate present per region in the Philippines from the year 2022. The worker's daily minimum wage included in this dataframe is subdivided into three: Non-Agriculture such as those who are in construction, factories and such. Agriculture Plantation are those workers employed in any agricultural enterprise within the 24 hectares area like farmers and finally, Agriculture Non-Plantation are those agricultural workers beyond the said area.



```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/kryzlerahne/IT3205_PhilippineDataAlliance/main/Datasets/Current%20Real%20MW%20February%202022.csv'
df = pd.read_csv(url)
newdf = df.drop([0,1,2], axis=0)
newdf.rename(columns={'Current Real Minimum Wage Based on February 2022 CPI (2018=100)': 'Region','Unnamed: 1': 'NonAgriculture', 'Unnamed: 2': 'AgriculturePlantation', 'Unnamed: 3': 'AgricultureNonPlantation'}, inplace=True)

display(newdf.iloc[0:])
```

```python
#converting datatype
newdf['NonAgriculture']= newdf.NonAgriculture.astype(float)
newdf['AgriculturePlantation']= newdf.AgriculturePlantation.astype(float)
newdf['AgricultureNonPlantation']= newdf.AgricultureNonPlantation.astype(float)

#getting all values from the dataframe/csv
Region = newdf["Region"]
NonAgriculture = newdf["NonAgriculture"]
AgriculturePlantation = newdf["AgriculturePlantation"]
AgricultureNonPlantation = newdf["AgricultureNonPlantation"]

#size of the bar graph
f = plt.figure()
f.set_figwidth(23)
f.set_figheight(10)
plt.rcParams['axes.facecolor'] = '#F2F2F2'

#Setting labels
plt.ylabel("Minimum Wage 2022" , fontsize=15)
plt.xlabel("Regions", fontsize=15)
plt.title("\n Summary of Daily Minimum Wage Rate in the Philippines (2022)",  fontweight="bold" , fontsize=20)

#creating visualizations 
xpos = np.arange(len(Region))
widthSize=0.25
plt.bar(xpos, NonAgriculture, color="#FAAA17", width = widthSize, label='Non-Agriculture')
plt.bar(xpos+0.3, AgriculturePlantation, color="#AFD43F", width = widthSize, label='Agriculture Plantation')
plt.bar(xpos+0.6, AgricultureNonPlantation, color="#EBC590", width = widthSize, label='Agriculture Non-Plantation')
plt.xticks(xpos+0.3, ('NCR', 'CAR', 'I', 'II','III','IV-A', 'IV-B', 'V', 'VI','VII','VIII','IX', 'X','XI', 'XII', 'CARAGA','BARMM'), fontsize=10)
plt.yticks(fontsize=12)
plt.legend()
#plt.savefig('wage.png', transparent=True)
plt.show()
```

## Interpretation: 
<p align=justify>The bar graph presented is entitled the "Summary of Philippine Daily Minimum Wage Rates (2022)", which showcases the minimum wage in the Philippines per region based on specific categories. As shown in the diagram, the minimum wage in the Philippines varies according to its region. The y-axis is represented by the minimum wage; on the other hand, the x-axis is composed of the region. The bar graphs are categorized by their color. The green bars indicate non-agriculture; the yellow bars are for agriculture plantations; and lastly, the red bars indicate agriculture non-plantation. Based on the graph, the National Capital Region (NCR) has the highest rate of minimum wage across the country in all the categories mentioned. In contrast, the Bicol Region has the lowest. Furthermore, the Ilocos Region (Region I) has the lowest agricultural minimum wage rate, while the Bicol Region (Region V) has the lowest non-agriculture minimum wage rate. There are three regions that have an equal number of minimum wages in all categories: the MIMAROPA Region (Region IV-B), the Bicol Region (Region V), and the Caraga Administrative Region (CAR).

According to the study by Cacnio (2017), the variations in minimum wages and salary increases among regions reflect changes in the cost of living in various locations. Also taken into consideration are socio-economic factors such as investment development and job creation.

# Current Minimum Wage and Inflation Rate per Region (2022)
*Source: Department of Labor and Employment's National Wages and Productivity Commission*


*   <p align=justify>The following provides information on the current minimum wage and inflation rate per region present in the Philippines from the year 2022. This dataset helps the researchers in identifying how does inflation rate greatly affect the current minimum wage being mandated in every region, and thus, provides us an answer as to why minimum wages vary from one area to the other area.

```python
#Dataframe
url = 'https://raw.githubusercontent.com/kryzlerahne/IT3205_PhilippineDataAlliance/main/Datasets/Daily%20Minimum%20Wage%20By%20Region.csv'
df = pd.read_csv(url)
display(df.iloc[0:])
```

```python
#getting all values from the dataframe/csv
Region = df["Region"]
currentWage = df["Current Wage"]
inflation = df["Inflation Rate (%)"]

#changing the size of the entire graph
f = plt.figure()
f.set_figwidth(23)
f.set_figheight(7)

#Changing the background colors of the graph
plt.rcParams['axes.facecolor'] = '#F2F2F2'

#Plot
plt.grid(color = 'green', linestyle = '--', linewidth = 0.3)
plt.scatter(Region, currentWage, c=inflation, s=300, cmap='autumn')
plt.title("Current Minimum Wage Vs Inflation Rate Per Region", fontweight="bold")
plt.xlabel("Region")
plt.ylabel("Current Minimum Wage (in Pesos)")

#Using of Colorbar to present the third variable (Inflation Rate)
plt.colorbar(label='Inflation Rate')
plt.show()
```

## Interpretation:

<p align=justify>A scatter plot diagram was devised in the visualization for the"Current Minimum Wage vs. Inflation Rate per Region," which also has a color bar. The plot in the diagram depicts the minimum wage (in pesos) of a region. The labels on the x-axis mark the regions in the Philippines. The color bar on the right side of the figure signifies the inflation rate. This figure shows that the NCR has the highest minimum wage at 537 pesos per day, with a moderate inflation rate of 4.4%. Conversely, Region V has the lowest rate for its minimum wage, at 263.61 pesos, and an inflation rate of 4.2%. It is also evident that the three regions that have the highest inflation rate of 6.3%, 6.1%, and 6.1% are CAR, Region III, and Region VIII, respectively. On the other note, BARMM has the lowest inflation rate with 2.1%.

As cited in the study entitled "The Price Effect of Minimum Wage: Evidence from the Philippines" by Cacnio (2017) indicated that inflation and the minimum wage have an impact on the development of society. A higher minimum wage will eventually lead to rising production costs. In which the industry may either pass to the consumers by raising the prices of the goods and services it offers or alter the production needs and equipments.

# Unemployment Rate in the Philippines (2021-2022) 
*Source: Philippine Statistics Authority*


*   <p align=justify>The following provides information on the current unemployment rate present in the Philippines from year 2021-2022 that will be used for analyzing one of the country's major concerns and exploring if these are somehow dependent on the minimum wage that this country have right now.


```python
#Dataframe
url = 'https://raw.githubusercontent.com/kryzlerahne/IT3205_PhilippineDataAlliance/main/Datasets/Unemployment%20Rate.csv'
df = pd.read_csv(url)
display(df.iloc[0:])
```

```python
#Visualization
#getting all values from the dataframe/csv
Date = df["Date"]
Percentage = df["Percentage"]

#changing the size of the entire graph
f = plt.figure(tight_layout=True)
f.set_figwidth(13)
f.set_figheight(7)

#Changing the background color of the graph
plt.rcParams['axes.facecolor'] = '#F2F2F2'

#Creating plots
plt.title("\n Unemployment Rate in Philippines", fontweight="bold")
plt.xlabel("Reference Date")
plt.ylabel("Percentage")
plt.grid(color = 'white', linestyle = '--', linewidth = 1.5)
plt.plot(Date, Percentage, 'o-', linewidth=2)
plt.legend(labels=['Unemployment Rate'])
plt.show() 
```

The number of unemployed was at 2.88 million, down from 3.44 million in March 2021. Meanwhile, the number of employed came in at 46.98 million, up from 4533 million the year before. Among employed persons, workers in the services sector made up 57.4 percent of the total, followed by those in the agriculture sector (25.2 percent) and industry (17.4 percent). The labor force participation rate edged up to 65.4 percent, the highest since the start of the pandemic, from 65.0 in March 2021. In February, the jobless rate was at 6.4 percent.
