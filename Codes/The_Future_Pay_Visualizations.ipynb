# Summary of Daily Minimum Wage Rate in the Philippines (2022)
*Source: Philippine Statistics Authority*


* <p align=justify> The following data provides information on the current daily minimum wage rate present per region in the Philippines from the year 2022. The worker's daily minimum wage included in this dataframe is subdivided into three: Non-Agriculture such as those who are in construction, factories and such. Agriculture Plantation are those workers employed in any agricultural enterprise within the 24 hectares area like farmers and finally, Agriculture Non-Plantation are those agricultural workers beyond the said area.



```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns
import plotly as py
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
import plotly.graph_objs as go

#Changing of fonts
plt.rcParams["font.family"] = "serif"
```

```python
url = 'https://raw.githubusercontent.com/kryzlerahne/IT3205_PhilippineDataAlliance/main/Datasets/Current%20Real%20MW%20February%202022.csv'
df = pd.read_csv(url)
newdf = df.drop([0,1,2], axis=0)
newdf.rename(columns={'Current Real Minimum Wage Based on February 2022 CPI (2018=100)': 'Region','Unnamed: 1': 'NonAgriculture', 'Unnamed: 2': 'AgriculturePlantation', 'Unnamed: 3': 'AgricultureNonPlantation'}, inplace=True)

display(newdf.iloc[0:])
```

```python
#converting datatype
newdf['NonAgriculture']= newdf.NonAgriculture.astype(float)
newdf['AgriculturePlantation']= newdf.AgriculturePlantation.astype(float)
newdf['AgricultureNonPlantation']= newdf.AgricultureNonPlantation.astype(float)

#getting all values from the dataframe/csv
Region = newdf["Region"]
NonAgriculture = newdf["NonAgriculture"]
AgriculturePlantation = newdf["AgriculturePlantation"]
AgricultureNonPlantation = newdf["AgricultureNonPlantation"]

#size of the bar graph
f = plt.figure()
f.set_figwidth(23)
f.set_figheight(10)
plt.rcParams['axes.facecolor'] = '#F2F2F2'

#Setting labels
plt.ylabel("Minimum Wage 2022" , fontsize=15)
plt.xlabel("Regions", fontsize=15)
plt.title("\n Summary of Daily Minimum Wage Rate in the Philippines (2022)",  fontweight="bold" , fontsize=20)

#creating visualizations 
xpos = np.arange(len(Region))
widthSize=0.25
plt.bar(xpos, NonAgriculture, color="#FAAA17", width = widthSize, label='Non-Agriculture')
plt.bar(xpos+0.3, AgriculturePlantation, color="#AFD43F", width = widthSize, label='Agriculture Plantation')
plt.bar(xpos+0.6, AgricultureNonPlantation, color="#EBC590", width = widthSize, label='Agriculture Non-Plantation')
plt.xticks(xpos+0.3, ('NCR', 'CAR', 'I', 'II','III','IV-A', 'IV-B', 'V', 'VI','VII','VIII','IX', 'X','XI', 'XII', 'CARAGA','BARMM'), fontsize=12)
plt.yticks(fontsize=12)
plt.legend()
#Saving visualizations in TIFF file with 300 dpi
plt.savefig('SummaryWage.tiff', transparent=True, dpi=300)
plt.show()
```

## Interpretation: 
<p align=justify>The bar graph presented is entitled the "Summary of Philippine Daily Minimum Wage Rates (2022)", which showcases the minimum wage in the Philippines per region based on specific categories. As shown in the diagram, the minimum wage in the Philippines varies according to its region. The y-axis is represented by the minimum wage; on the other hand, the x-axis is composed of the region. The bar graphs are categorized by their color. The green bars indicate non-agriculture; the yellow bars are for agriculture plantations; and lastly, the red bars indicate agriculture non-plantation. Based on the graph, the National Capital Region (NCR) has the highest rate of minimum wage across the country in all the categories mentioned. In contrast, the Bicol Region has the lowest. Furthermore, the Ilocos Region (Region I) has the lowest agricultural minimum wage rate, while the Bicol Region (Region V) has the lowest non-agriculture minimum wage rate. There are three regions that have an equal number of minimum wages in all categories: the MIMAROPA Region (Region IV-B), the Bicol Region (Region V), and the Caraga Administrative Region (CAR).

## Analysis:
<b>The daily minimum wage of those non-agricultural workers tends to be higher than those two other categories of Filipino workers. While NCR ranked first and has the highest minimum wage among all regions in the Philippines. </b>
<p align=justify> According to the study by Cacnio (2017), the variations in minimum wages and salary increases among regions reflect changes in the cost of living in various locations. Also taken into consideration are socio-economic factors such as investment development and job creation.

# Current Minimum Wage and Inflation Rate per Region (2022)
*Source: Department of Labor and Employment's National Wages and Productivity Commission*


*   <p align=justify>The following provides information on the current minimum wage and inflation rate per region present in the Philippines from the year 2022. This dataset helps the researchers in identifying how does inflation rate greatly affect the current minimum wage being mandated in every region, and thus, provides us an answer as to why minimum wages vary from one area to the other area.

```python
#Dataframe
url = 'https://raw.githubusercontent.com/kryzlerahne/IT3205_PhilippineDataAlliance/main/Datasets/Daily%20Minimum%20Wage%20By%20Region.csv'
df = pd.read_csv(url)
display(df.iloc[0:])
```

```python
#getting all values from the dataframe/csv
Region = df["Region"]
currentWage = df["Current Wage"]
inflation = df["Inflation Rate (%)"]

#changing the size of the entire graph
f = plt.figure()
f.set_figwidth(23)
f.set_figheight(7)

#Changing the background colors of the graph
plt.rcParams['axes.facecolor'] = '#F2F2F2'

#Plot
plt.grid(color = 'green', linestyle = '--', linewidth = 0.3)
plt.scatter(Region, currentWage, c=inflation, s=300, cmap='autumn')
plt.title("Current Minimum Wage Vs Inflation Rate Per Region", fontweight="bold" , fontsize=20)
plt.xlabel("Region", fontsize=12)
plt.ylabel("Current Minimum Wage (in Pesos)", fontsize=12)

#Using of Colorbar to present the third variable (Inflation Rate)
plt.colorbar(label='Inflation Rate')
#Saving visualizations in TIFF file with 300 dpi
plt.savefig('CurrentWageVsInflation.tiff', transparent=True, dpi=300)
plt.show()
```

## Interpretation:

<p align=justify>A scatter plot diagram was devised in the visualization for the"Current Minimum Wage vs. Inflation Rate per Region," which also has a color bar. The plot in the diagram depicts the minimum wage (in pesos) of a region. The labels on the x-axis mark the regions in the Philippines. The color bar on the right side of the figure signifies the inflation rate. This figure shows that the NCR has the highest minimum wage at 537 pesos per day, with a moderate inflation rate of 4.4%. Conversely, Region V has the lowest rate for its minimum wage, at 263.61 pesos, and an inflation rate of 4.2%. It is also evident that the three regions that have the highest inflation rate of 6.3%, 6.1%, and 6.1% are CAR, Region III, and Region VIII, respectively. On the other note, BARMM has the lowest inflation rate with 2.1%.

## Analysis: 
<b> NCR having the highest minimum wage has relatively mild inflation rate of 4.4%, whereas CAR have the worst inflation rate of all regions, while having a low minimum pay in comparison to others. We may assume that measures and actions should be taken, to assist Filipinos in living within their needs to survive while the inflation rate continues to rise.  </b>

As cited in the study entitled "The Price Effect of Minimum Wage: Evidence from the Philippines" by Cacnio (2017) indicated that inflation and the minimum wage have an impact on the development of society. A higher minimum wage will eventually lead to rising production costs. In which the industry may either pass to the consumers by raising the prices of the goods and services it offers or alter the production needs and equipments.

# Inflation Rate in the Philippines (2019-2022)
*Source: Philippine Statistics Authority*


*  <p align = justify> The following data provides information on the current inflation rate present in the Philippines from year 2019-2022 which will allow us to measure if this affects the minimum wages of the Filipino workers.


```python
url = 'https://raw.githubusercontent.com/kryzlerahne/IT3205_PhilippineDataAlliance/main/Datasets/Inflation%20Rates%20in%20PH.csv'
df = pd.read_csv(url)
new_df = df.drop([0,1], axis=0)
new_df.rename(columns={'Year on Year Inflation Rates in the Philippines (January 2019 - April 2022)': 'Month','Unnamed: 1': 'InflationRate2019', 'Unnamed: 2': 'InflationRate2020', 'Unnamed: 3': 'InflationRate2021', 'Unnamed: 4': 'InflationRate2022'}, inplace=True)

#data cleaning
new_df['InflationRate2022'] = new_df['InflationRate2022'].fillna(0)

display(new_df)
```

```python
#converting datatype  
new_df['InflationRate2019']= new_df.InflationRate2019.astype(float)

#getting all values from the dataframe/csv
Month = new_df["Month"]
InflationRate2019 = new_df["InflationRate2019"]
InflationRate2020 = new_df["InflationRate2020"]
InflationRate2021 = new_df["InflationRate2021"]
InflationRate2022 = new_df["InflationRate2022"]

#changing the size of the entire graph
f = plt.figure(tight_layout=True)
f.set_figwidth(23)
f.set_figheight(10)

#Changing the background color of the graph
plt.rcParams['axes.facecolor'] = '#F2F2F2'

#Creating plots
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title("Inflation Rate in Philippines 2019-2022", fontweight="bold", fontsize=20)
plt.xlabel("Months", fontsize=15)
plt.ylabel("Inflation Rate", fontsize=15)
plt.grid(color = 'white', linestyle = '--', linewidth = 0.8)
plt.plot(Month, InflationRate2019, 'o--', color="#E33B3A", label="2019", lw=2.4)
plt.plot(Month, InflationRate2020, 'o--', color="#AFD43F", label="2020", lw=2.4)
plt.plot(Month, InflationRate2021, 'o--', color="#84B100", label="2021", lw=2.4)
plt.plot(Month, InflationRate2022, 'o--', color="#FAAA17", label="2022", lw=2.4)
plt.legend(title='Inflation Rate')
#Saving visualizations in TIFF file with 300 dpi
plt.savefig('InflationRate.tiff', transparent=True, dpi=300)
plt.show() 
```

## Interpretation:
<p align=justify>The line graph shown is labeled "Inflation Rate in the Philippines (2019-2022)," and it depicts the current inflation rate in the Philippines from 2019 to 2022, allowing us to determine if this impacts the minimum wages of Filipino employees. The inflation rate in the Philippines varies dramatically each year, as seen in the diagram. The inflation rate is depicted on the y-axis, while the months are represented on the x-axis. Color codes are used to classify the line graph. The red line represents the year 2019, the light-green line represents the year 2020, the dark-green line represents the year 2021, and the yellow line represents the year 2022. According to the graph, the year 2019 began with one of the highest rates of inflation, 4.4 percent, but then gradually drops throughout the year, reaching the lowest rate of inflation, 0.5 percent, in September. The largest rates of inflation may be witnessed between 2021 and 2022, ranging from 3 to 4.9 percent, which is 10 times the lowest rate of inflation. In April 2022, the annual inflation rate in the Philippines increased to 4.9 percent from 3.0 percent in February.

## Analysis:
<b> The inflation rate in the Philippines continues to rise up to the present year, specifically in month of April, compared to its past years. </b>

According to the Philippine Statistics Authority (2022), May 5 represents a 0.9-percentage-point increase from the 4 percent reported in March, with pump prices of petroleum commodities continuing to climb owing to the Russian-Ukrainian war. It is also the first time that the inflation rate has beyond the government's planned range of 2% to 4% in 2022.

# Poverty Statistics
*Source: Philippine Statistics Authority*


*   The following provides information on the current poverty statistics present in the Philippines from years 2018 & 2021, which is one of the factors being monitored by the researchers relating as one of the potential effect on increasing the minimum wage as recommended.


```python
url = 'https://raw.githubusercontent.com/kryzlerahne/IT3205_PhilippineDataAlliance/main/Datasets/Poverty%20Incidence%20Rate.csv'
df = pd.read_csv(url)
display(df)
```

```python
#getting all values from the dataframe/csv
Region = df["Region"]
PI2018 = df["Poverty Incidence 2018"]
PI2021 = df["Poverty Incidence 2021"]

#size of the entire graph
f = plt.figure()
f.set_figwidth(23)
f.set_figheight(10)

#creating visualizations
plt.xlabel("Regions", fontsize=15)
plt.ylabel("Poverty Instance Rate", fontsize=15)
plt.title("\n Poverty Rate in the Philippines", fontweight="bold", fontsize=20)

xpos = np.arange(len(Region))
widthSize=0.3
plt.bar(xpos, PI2018,width= widthSize, label='Poverty Incidence 2018', color = '#FAAA17')
plt.bar(xpos+0.3, PI2021, color="#84B100", width= widthSize, label='Poverty Incidence 2021')
plt.yticks(fontsize=12)
plt.xticks(xpos+0.1, ('NCR', 'CAR', 'I', 'II','III','IV-A', 'IV-B', 'V', 'VI','VII','VIII','IX', 'X','XI', 'XII', 'CARAGA','BARMM'), fontsize=13)
plt.legend()
#Saving visualizations in TIFF file with 300 dpi
plt.savefig('PovertyRate.tiff', transparent=True, dpi=300)
plt.show() 
```

## Interpretation:
As reported by the Philippine Statistics Authority (PSA) today, the poverty incidence among the population increased to 23.7 percent during the first half of 2021 from 21.1 percent in the same period of 2018. This translates to 3.9 million more Filipinos living in poverty. Regions with stricter quarantines tended to see larger increases in poverty compared to regions under less stringent quarantines. For instance, the poverty incidence in Regions III, IV-A, and VII increased by over four percentage points. The poverty incidence in the National Capital Region also increased by around 1.2 percentage points.

On the other hand, five regions with lesser quarantine restrictions recorded lower poverty. These are BARMM, and Regions VIII, IX, XI, and the Cordillera Administrative Region. # Linear Regression with Correlation Coefficients
### Minimum Wage Vs. Poverty

```python
#Linear Regression
snsplot = sns.lmplot(x="CurrentWage", y="PovertyRate", data=df)
plt.title("\n Relationship between Minimum Wage and Poverty", fontweight="bold")
#Saving visualizations in TIFF file with 300 dpi
snsplot.savefig('LinearPoverty.tiff', transparent=True, dpi=300)
```
