# Labor Force Employment (2021-2022)
*Source: Philippine Statistics Authority*


*   This will provide high frequency statistical information on the Philippine labor market to better understand the current situation of employment and the economy of the Philippines during unprecedented event such as the Corana Virus of 2019 (COVID-19) pandemic.




```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/kryzlerahne/IT3205_PhilippineDataAlliance/main/Datasets/Labor%20Force%20Employment%20(2021-2022).csv'
df = pd.read_csv(url)
display(df.iloc[0:, 0:3])
```

```python
#getting all the values stored in the dataframe/csv
Date = df["Date"]
People = df["Thousands of People"]

#changing the size of visualization
f = plt.figure()
f.set_figwidth(13)
f.set_figheight(7)

#creating the plot
plt.plot(Date, People)
plt.title("Labor Force Employment in the Philippines (2021-2022)", fontweight="bold")
plt.xlabel("Reference Date")
plt.ylabel("Number of Population")
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.show() 
```

# Unemployment Rate in the Philippines (2021-2022) 
*Source: Philippine Statistics Authority*


*   The following provides information on the current unemployment rate present in the Philippines from year 2021-2022.


```python
url = 'https://raw.githubusercontent.com/kryzlerahne/IT3205_PhilippineDataAlliance/main/Datasets/Unemployment%20Rate.csv'
df = pd.read_csv(url)
display(df.iloc[0:])
```

# Summary of Daily Minimum Wage Rate in the Philippines (2022)
*Source: Philippine Statistics Authority*

```python
url = 'https://raw.githubusercontent.com/kryzlerahne/IT3205_PhilippineDataAlliance/main/Datasets/Current%20Real%20MW%20February%202022.csv'
df = pd.read_csv(url)
newdf = df.drop([0,1,2], axis=0)
newdf.rename(columns={'Current Real Minimum Wage Based on February 2022 CPI (2018=100)': 'Region','Unnamed: 1': 'NonAgriculture', 'Unnamed: 2': 'AgriculturePlantation', 'Unnamed: 3': 'AgricultureNonPlantation'}, inplace=True)

display(newdf.iloc[0:])
```

```python
#converting datatype
newdf['NonAgriculture']= newdf.NonAgriculture.astype(float)
newdf['AgriculturePlantation']= newdf.AgriculturePlantation.astype(float)
newdf['AgricultureNonPlantation']= newdf.AgricultureNonPlantation.astype(float)
```

```python
#getting all values from the dataframe/csv

Region = newdf["Region"]
NonAgriculture = newdf["NonAgriculture"]
AgriculturePlantation = newdf["AgriculturePlantation"]
AgricultureNonPlantation = newdf["AgricultureNonPlantation"]

#size of the entire graph
f = plt.figure()
f.set_figwidth(25)
f.set_figheight(10)

#creating visualizations
plt.ylabel("Minimum Wage 2022" , fontsize=15)
plt.xlabel("Regions", fontsize=15)
plt.title("Summary of Daily Minimum Wage Rate in the Philippines (2022)", fontweight="bold" , fontsize=20)

xpos = np.arange(len(Region))
barTry=0.25
plt.bar(xpos, NonAgriculture, color="blue", width = barTry, label='Non-Agriculture')
plt.bar(xpos+0.3, AgriculturePlantation, color="hotpink", width = barTry , label='Agriculture Plantation')
plt.bar(xpos+0.6, AgricultureNonPlantation, color="yellow", width = barTry , label='Agriculture Non-Plantation')

plt.xticks(xpos+0.3, ('NCR', 'CAR', 'I', 'II','III','IV-A', 'IV-B', 'V', 'VI','VII','VIII','IX', 'X','XI', 'XII', 'CARAGA','BARMM'), fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```
